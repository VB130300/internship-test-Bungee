```python
import pandas as pd
df = pd.read_csv("/content/main.csv")
df.head(30)
```

TO KNOW THE MEAN VALUE OF EACH PRODUCT

```python
df.groupby('product_description')['price'].mean()
```

INTRODUING A NEW COLUMN WHICH HAS THE VALUES BASED ON THE MEAN OF ALL THE PRODUCTS

```python
df["price_new"] = df['price'].fillna(df.groupby('product_description')['price'].transform("mean"))
```

```python
df[df["price"].isna()].head()
```

```python
df['price_new'].isnull().sum()
```
